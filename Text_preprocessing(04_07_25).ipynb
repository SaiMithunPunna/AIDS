```python
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
```

```python


def preprocess(text):
  text = word_tokenize(text.lower())
  return text
```

```python
messages['tokens'] = messages['message'].apply(preprocess)
```

```python
def lematize(tokens):
  lemmatized_tokens = [WordNetLemmatizer().lemmatize(word) for word in tokens]
  return ' '.join(lemmatized_tokens)
```

```python
messages['lematized']=messages['tokens'].apply(lematize)
```

```python
def clean_text(tokens):
  port = PorterStemmer()
  words = [port.stem(word) for word in tokens if word not in set(stopwords.words('english'))]
  return ' '.join(words)

messages['stemming'] = messages['tokens'].apply(clean_text)
```

```python
corpus=[]
for i in range(0 , len(messages)):
  port=PorterStemmer()
  review=re.sub('[^a-zA-Z]' , ' ' , messages['message'][i])
  review=review.lower()
  review=review.split()

  review=[port.stem(word) for word in review if not word in stopwords.words('english')]
  review=' '.join(review)
  corpus.append(review)
```
