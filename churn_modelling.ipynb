```python
import pandas as pd
import seaborn as sns
import numpy as np
from sklearn.preprocessing import MinMaxScaler , LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('/content/drive/MyDrive/Colab Datasets/1_Churn_Modelling(in).csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
le=LabelEncoder()
```

```python
categorical_columns=['Geography','Gender']
```

```python
for col in categorical_columns:
  df[col]=le.fit_transform(df[col])
```

```python
df
```

```python
df.drop_duplicates()
```

```python
Q1=df[['CreditScore' ,'Geography' , 'Gender' , 'Age' , 'Tenure', 'Balance','NumOfProducts','HasCrCard','IsActiveMember','EstimatedSalary']].quantile(0.25)
Q3=df[['CreditScore' ,'Geography' , 'Gender' , 'Age' , 'Tenure', 'Balance','NumOfProducts','HasCrCard','IsActiveMember','EstimatedSalary']].quantile(0.75)
IQR=Q3-Q1
```

```python
rem_out=df[~(((df[['CreditScore' ,'Geography' , 'Gender' , 'Age' , 'Tenure', 'Balance','NumOfProducts','HasCrCard','IsActiveMember','EstimatedSalary']] < (Q1-1.5*IQR)) | (df[['CreditScore' ,'Geography' , 'Gender' , 'Age' , 'Tenure', 'Balance','NumOfProducts','HasCrCard','IsActiveMember','EstimatedSalary']]>(Q3+1.5 *IQR)))).any(axis=1)]
display(rem_out.head())
```

```python
df=rem_out
```

```python
df
```

```python
sc=MinMaxScaler()
```

```python
df[['CreditScore' ,'Geography' , 'Gender' , 'Age' , 'Tenure', 'Balance','NumOfProducts','HasCrCard','IsActiveMember']] =sc.fit_transform(df[['CreditScore' ,'Geography' , 'Gender' , 'Age' , 'Tenure', 'Balance','NumOfProducts','HasCrCard','IsActiveMember']])
```

```python
df
```

```python
X=df[['CreditScore' ,'Geography' , 'Gender' , 'Age' , 'Tenure', 'Balance','NumOfProducts','HasCrCard','IsActiveMember']]
y=df['EstimatedSalary']
```

```python
X
```

```python
y
```

```python
X_train , X_test , y_train , y_test=train_test_split(X,y,random_state=42)
```

#Modeling


```python
from keras import Sequential
from keras.layers import Dense
```

```python
model=Sequential()

model.add(Dense(units=128 ,input_shape=(X_train.shape[1],),activation='relu'))
model.add(Dense(units=64,kernel_initializer='he_uniform',activation='relu'))
model.add(Dense(units=32 ,kernel_initializer='he_uniform',activation='relu'))
model.add(Dense(units=1 ,kernel_initializer='he_uniform'))
```

```python
model.compile(loss='mean_squared_error' , optimizer='adam' , metrics=['mse'])
model.fit(X_train,y_train,batch_size=32 , epochs=10,verbose=1,validation_split=0.3)
```

```python
y_pred = model.predict(X_test)
print("Sample predictions vs actual:")
for i in range(5):
    print(f"Predicted: {y_pred[i][0]:.2f}, Actual: {y_test.iloc[i]:.2f}")
```
