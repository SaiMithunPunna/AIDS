```python
import seaborn as sns
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import LabelEncoder , StandardScaler
```

```python
df=sns.load_dataset('taxis')
```

```python

df
```

```python
df.drop_duplicates()
```

```python
sns.boxplot(df)
```

```python
import numpy as np
```

```python
le=LabelEncoder()
df['payment']=le.fit_transform(df['payment'])
df.head()
```

```python
Q1=df[['distance' ,'fare' , 'tip' , 'tolls' , 'total']].quantile(0.25)
Q3=df[['distance' ,'fare' , 'tip' , 'tolls' , 'total']].quantile(0.75)
IQR=Q3-Q1
```

```python
rem_out=df[~(((df[['distance' ,'fare' , 'tip' , 'tolls' , 'total']] < (Q1-1.5*IQR)) | (df[['distance' ,'fare' , 'tip' , 'tolls' , 'total']]>(Q3+1.5 *IQR)))).any(axis=1)]
display(rem_out.head())
```

```python
df=rem_out
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
sc=MinMaxScaler()
df[['passengers','distance', 'fare' , 'tip' , 'tolls' , 'payment' ]] =sc.fit_transform(df[['passengers','distance', 'fare' , 'tip' , 'tolls' , 'payment' ]])
```

```python
df
```

```python
X=df[['passengers','distance', 'fare' , 'tip' , 'tolls' , 'payment' ]]
y=df['total']
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test=train_test_split(X,y,random_state=42)
```

# Model Building

```python
from keras import Sequential
from keras.layers import Dense


```

```python
model=Sequential()

model.add(Dense(units=128 ,input_shape=(X_train.shape[1],),activation='relu'))
model.add(Dense(units=64,kernel_initializer='he_uniform',activation='relu'))
model.add(Dense(units=32 ,kernel_initializer='he_uniform',activation='relu'))
model.add(Dense(units=1 ,kernel_initializer='he_uniform'))
```

```python
model.compile(loss='mean_squared_error' , optimizer='adam' , metrics=['mse'])
model.fit(X_train,y_train,batch_size=32 , epochs=10,verbose=1,validation_split=0.3)
```

```python
y_pred = model.predict(X_test)
print("Sample predictions vs actual:")
for i in range(5):
    print(f"Predicted: {y_pred[i][0]:.2f}, Actual: {y_test.iloc[i]:.2f}")
```
