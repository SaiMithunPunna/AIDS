```python

import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/drive/MyDrive/Colab Datasets/marketing_campaign1.csv')
df.head()
```

```python
df['Income'].fillna(df['Income'].mean() , inplace=True)
```

```python
df.isnull().sum()
df
```

```python

numeric_data = df.select_dtypes(include=[np.number])
```

```python

# Perform K-means clustering with K=2
kmeans = KMeans(n_clusters=2, random_state=0).fit(numeric_data)
```

```python
# Get the cluster centers (centroids) and labels
centroids = kmeans.cluster_centers_
labels = kmeans.labels_
```

```python
# Print the final centroids and cluster assignments
print("Final Centroids (Cluster Centers):")
print(centroids)
```

```python
print("\nCluster Assignments:")
for i, label in enumerate(labels):
    print(f"Point {chr(65+i)}: Cluster {label+1}")
```

```python
X = numeric_data
wcss = []

# Test different values of k (number of clusters)
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)  # inertia_ is the wcss

# Plot the elbow method graph
plt.plot(range(1, 11), wcss, marker='o', linestyle='-')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.title('The Elbow Method')
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
z=linkage(X, 'ward')
dendrogram(z)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Data point')
plt.ylabel('Distance')
plt.show()
```
